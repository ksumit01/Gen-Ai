## Rock,Paper, Scissor Game

```python
import random

userScore = 0
computerScore = 0

while True:
    choices = ["rock", "paper", "scissors"]
    computer_choice = random.choice(choices)

    while True:
        user_input = input("Enter your choice (rock, paper, or scissors) or 'quit' to exit: ").lower()

        if user_input == 'quit':
            break  # Exit the inner loop if the user wants to quit the game.

        if user_input not in choices:
            print("Invalid choice. Please choose 'rock', 'paper', or 'scissors'.")
        else:
            break  # Exit the inner loop if valid input is provided.

    if user_input == 'quit':
        break  # Exit the outer loop if the user wants to quit the game.

    print("\nComputer's choice:", computer_choice)

    if computer_choice == "rock" and (user_input == "paper" or user_input == "scissors"):
        userScore += 1
        print("\nYou win! Paper beats Rock.")
    elif computer_choice == "paper" and (user_input == "rock" or user_input == "scissors"):
        computerScore += 1
        print("\nComputer wins! Paper beats Rock.")
    elif computer_choice == "scissors" and (user_input == "rock" or user_input == "paper"):
        userScore += 1
        print("\nYou win! Rock beats Scissors.")
    elif computer_choice == "rock" and user_input == "scissors":
        computerScore += 1
        print("\nComputer wins! Rock beats Scissors.")
    elif computer_choice == "scissors" and user_input == "paper":
        computerScore += 1
        print("\nComputer wins! Scissors beats Paper.")
    elif computer_choice == "paper" and user_input == "rock":
        userScore += 1
        print("\nYou win! Paper beats Rock.")
    else:
        print("\nIt's a draw!")

    print("\nYour score:", userScore)
    print("Computer's score:", computerScore)
    print("\n" + "-" * 30)  # Line separator

# The loop will keep running until the user enters 'quit'.
print("\nThanks for playing!")
```
